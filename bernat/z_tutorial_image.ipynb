# Tutorial: Images

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
```

## Import and Show an image

```python
img = mpimg.imread('images/sampletext2.png')

print(img)

plt.imshow(img)
plt.show()

lum_img = img[:, :, 0]

plt.imshow(lum_img, cmap='hot')
plt.colorbar()
plt.show()
```

## Histogram

```python
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

## Contrast

```python
fig = plt.figure()

ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(lum_img, cmap='hot')
ax.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img, cmap='hot')
imgplot.set_clim(0.0, 0.7)
ax.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

plt.show()
```

## Import .JPG

```python
h = Image.open('images/h_0001.jpg')
realsize=h.size
print(realsize)
factor=4
newsize=tuple(int(ti/factor) for ti in realsize)
print(newsize)
h=h.resize(newsize,Image.ANTIALIAS)
h = np.array(h,dtype=float) #We convert it to an ARRAY.
h = h[:, :, 0] #For instance, at point (100,50) we had [0.608, 0.608, 0.608]. Now only have a single 0.608 value.
size_x = len(h[0])
size_y = len(h)
print(h.shape) #[Y,X]
print(size_x)
print(size_y)
print('h(t) imported and reduced correctly!')

plt.imshow(h, cmap='gray')
plt.show()
```

## Import .PNG

```python
h = Image.open('images/sampletext2.png')
realsize=h.size
print(realsize)
factor=2
newsize=tuple(int(ti/factor) for ti in realsize)
print(newsize)
h=h.resize(newsize,Image.ANTIALIAS)
h = np.array(h,dtype=float) #We convert it to an ARRAY.
h = h[:, :, 0] #One option to convert from RGB to grayscale is to get one color (R,G or B)
size_x = len(h[0])
size_y = len(h)
print(h.shape) #[Y,X]
print(size_x)
print(size_y)
print('h(t) imported and reduced correctly!')

print(h)
plt.imshow(h, cmap='gray')
plt.show()
```
