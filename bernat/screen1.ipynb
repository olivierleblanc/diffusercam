```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import math
from skimage import draw
from scipy import signal
from PIL import Image
```

```python
def import_capture(location,factor):
    img = Image.open(location)
    realsize=img.size
    newsize=tuple(int(ti/factor) for ti in realsize)
    img=img.resize(newsize,Image.ANTIALIAS)
    img = np.array(img,dtype=float)
    img = img[:, :, 0]
    return img

def import_capture_mean(location,factor):
    img = Image.open(location)
    realsize=img.size
    newsize=tuple(int(ti/factor) for ti in realsize)
    img=img.resize(newsize,Image.ANTIALIAS)
    img = np.array(img,dtype=float)
    img = img[:, :, 0]
    img -= img.mean()
    return img
```

```python
#Import image
h = import_capture('from_camera/test03/dist_00000.png', 4)
h -= h.mean()

black = import_capture('from_camera/test05/black5.png', 4)
og_black = signal.fftconvolve(black,np.flip(h),mode='same')

point = import_capture('from_camera/test05/a1.png', 4)
og_point = signal.fftconvolve(point,np.flip(h),mode='same')

# plt.imshow(og_black)
# plt.colorbar(orientation='horizontal')
# plt.show()

# plt.imshow(og_point)
# plt.colorbar(orientation='horizontal')
# plt.show()

plt.imshow(og_point-og_black)
plt.colorbar(orientation='horizontal')
plt.show()
```
