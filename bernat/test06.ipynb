```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import math
from skimage import draw
from scipy import signal
from PIL import Image
```

```python
global_factor = 4

def image_open(name):
    location='from_camera/test06/'+name+'.png'
    image_im = Image.open(location)
    realsize=image_im.size
    newsize=tuple(int(ti/global_factor) for ti in realsize)
    image_im=image_im.resize(newsize,Image.ANTIALIAS)
    image_im = np.array(image_im,dtype=float)
    image_im = image_im[:, :, 0]
    return image_im
```

```python
h = image_open('a_psf')
h -= h.mean()
plt.imshow(h,cmap='gray')
plt.show()
```

# Part II - Arranged by symbol instead of method

```python
b_black1=image_open('b_black1')
c_black1=image_open('c_black1')

b_black10=image_open('b_black1')+image_open('b_black2')+image_open('b_black3')+image_open('b_black4')+image_open('b_black5')+image_open('b_black6')+image_open('b_black7')+image_open('b_black8')+image_open('b_black9')+image_open('b_black0')
c_black10=image_open('c_black1')+image_open('c_black2')+image_open('c_black3')+image_open('c_black4')+image_open('c_black5')+image_open('c_black6')+image_open('c_black7')+image_open('c_black8')+image_open('c_black9')+image_open('c_black0')
b_black10 /= 10
c_black10 /= 10

def estimation_b(name):
    imag=image_open(name)
    imag_hat=signal.fftconvolve(imag,np.flip(h),mode='same')
    imag_hat_1=signal.fftconvolve(imag-b_black1,np.flip(h),mode='same')
    imag_hat_10=signal.fftconvolve(imag-b_black10,np.flip(h),mode='same')
    fig = plt.figure(figsize=(25,25))
    fig.add_subplot(1, 4, 1)
    plt.imshow(imag,cmap='gray')
    plt.title(name)
    fig.add_subplot(1, 4, 2)
    plt.imshow(imag_hat,cmap='gray')
    fig.add_subplot(1, 4, 3)
    plt.imshow(imag_hat_1,cmap='gray')
    fig.add_subplot(1, 4, 4)
    plt.imshow(imag_hat_10,cmap='gray')
    plt.show()

def estimation_c(name):
    imag=image_open(name)
    imag_hat=signal.fftconvolve(imag,np.flip(h),mode='same')
    imag_hat_1=signal.fftconvolve(imag-c_black1,np.flip(h),mode='same')
    imag_hat_10=signal.fftconvolve(imag-c_black10,np.flip(h),mode='same')
    fig = plt.figure(figsize=(25,25))
    fig.add_subplot(1, 4, 1)
    plt.imshow(imag,cmap='gray')
    plt.title(name)
    fig.add_subplot(1, 4, 2)
    plt.imshow(imag_hat,cmap='gray')
    fig.add_subplot(1, 4, 3)
    plt.imshow(imag_hat_1,cmap='gray')
    fig.add_subplot(1, 4, 4)
    plt.imshow(imag_hat_10,cmap='gray')
    plt.show()
```

```python
estimation_b('b_dotcenter')
estimation_c('c_dotcenter')
estimation_b('b_bigdotcenter')
estimation_c('c_bigdotcenter')
estimation_b('b_dottopleft')
estimation_b('b_dottopright')
estimation_b('b_dotbottomleft')
estimation_b('b_dotbottomright')
estimation_b('b_horitzontal')
estimation_c('c_horitzontal')
estimation_b('b_vertical')
estimation_c('c_vertical')
estimation_b('b_diagonal1')
estimation_c('c_diagonal1')
estimation_b('b_diagonal2')
estimation_c('c_diagonal2')
estimation_b('b_lettera')
estimation_c('c_lettera')
estimation_b('b_happy')
estimation_c('c_happy')
```
