# Tutorial: Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

## One function

```python
x_array=np.array([],dtype=float)
y_array=np.array([],dtype=float)
for i in range(11):
    x=i
    y=x**2/10 #The function we want to plot: (x^2)/10
    x_array=np.insert(x_array,len(x_array),x)
    y_array=np.insert(y_array,len(y_array),y)
```

And we print the arrays and plot the results

```python
print(x_array)
print(y_array)

comb_array=np.append(x_array,y_array)
comb_array=np.resize(comb_array,(2,len(x_array)))
print(comb_array)

plt.plot(x_array,y_array)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('x^2/10')
plt.show()
```

## Two functions

```python
x_array=np.array([],dtype=float)
y1_array=np.array([],dtype=float)
y2_array=np.array([],dtype=float)
for i in range(21):
    x=i-10
    y1=x**2/10 #The function we want to plot: (x^2)/10
    y2=2*np.sin(x)+5
    x_array=np.insert(x_array,len(x_array),x)
    y1_array=np.insert(y1_array,len(y1_array),y1)
    y2_array=np.insert(y2_array,len(y2_array),y2)
```

```python
print(x_array)

plt.plot(x_array,y1_array,label='f1(x)')
plt.plot(x_array,y2_array,label='f2(x)')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('two diferent functions')
plt.legend()
plt.show()
```

## Smooth output

```python
x_array=np.array([],dtype=float)
y1_array=np.array([],dtype=float)
y2_array=np.array([],dtype=float)
for i in range(1001):
    x=i/20-25
    y1=np.sin(x)
    y2=np.sin(x)/x
    x_array=np.insert(x_array,len(x_array),x)
    y1_array=np.insert(y1_array,len(y1_array),y1)
    y2_array=np.insert(y2_array,len(y2_array),y2)
```

```python
print(x_array)

plt.plot(x_array,y1_array,label='f1(x)')
plt.plot(x_array,y2_array,label='f2(x)')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('two diferent functions')
plt.legend()
plt.show()
```

## Plotting a defined function

First, at the TOP, we define a function.

```python
def factorial_function(n):
    product = 1
    for ind in range(1, n+1):
        product = product * ind
    return product
```

```python
x_array=np.array([],dtype=float)
y_array=np.array([],dtype=float)
for i in range(7):
    x=i
    y=factorial_function(x) #The function we want to plot
    x_array=np.insert(x_array,len(x_array),x)
    y_array=np.insert(y_array,len(y_array),y)

#Plotting:
comb_array=np.append(x_array,y_array)
comb_array=np.resize(comb_array,(2,len(x_array)))
print(comb_array)

plt.plot(x_array,y_array)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('x^2/10')
plt.show()
```
