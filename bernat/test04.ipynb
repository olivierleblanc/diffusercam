```python
def fun_ratio(autocon,radi):
    [sy,sx]=autocon.shape
    max_=np.amax(autocon)
    [max_y,max_x]=np.where(autocon==max_)
    max_y=float(max_y)
    max_x=float(max_x)
    disk_mask = np.ones((sy, sx), dtype=float)
    rr, cc = draw.disk((max_y,max_x),radi)
    disk_mask[rr, cc] = 0
    autocon = np.maximum(autocon,np.ones(autocon.shape)) ### HERE !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    autocon_masked=disk_mask*autocon
    sum_nz=sum(sum(autocon_masked))
    number_nz=autocon.size-len(rr)
    average_nz=sum_nz/number_nz
    ratio_nz=max_/average_nz
    return ratio_nz

radii=2

rati_0200=fun_ratio(auto_0200,radii)
rati_0225=fun_ratio(auto_0225,radii)
rati_0250=fun_ratio(auto_0250,radii)
rati_0275=fun_ratio(auto_0275,radii)
rati_0300=fun_ratio(auto_0300,radii)
rati_0325=fun_ratio(auto_0325,radii)
rati_0350=fun_ratio(auto_0350,radii)
rati_0375=fun_ratio(auto_0375,radii)
rati_0400=fun_ratio(auto_0400,radii)
rati_0425=fun_ratio(auto_0425,radii)
rati_0450=fun_ratio(auto_0450,radii)
rati_0475=fun_ratio(auto_0475,radii)
rati_0500=fun_ratio(auto_0500,radii)
rati_0525=fun_ratio(auto_0525,radii)
rati_0550=fun_ratio(auto_0550,radii)
rati_0575=fun_ratio(auto_0575,radii)
rati_0600=fun_ratio(auto_0600,radii)
rati_0625=fun_ratio(auto_0625,radii)
rati_0650=fun_ratio(auto_0650,radii)
rati_0675=fun_ratio(auto_0675,radii)
rati_0700=fun_ratio(auto_0700,radii)
rati_0725=fun_ratio(auto_0725,radii)
rati_0750=fun_ratio(auto_0750,radii)
rati_0775=fun_ratio(auto_0775,radii)
rati_0800=fun_ratio(auto_0800,radii)
rati_0825=fun_ratio(auto_0825,radii)
rati_0850=fun_ratio(auto_0850,radii)
rati_0875=fun_ratio(auto_0875,radii)
rati_0900=fun_ratio(auto_0900,radii)
rati_0925=fun_ratio(auto_0925,radii)
rati_0950=fun_ratio(auto_0950,radii)
rati_0975=fun_ratio(auto_0975,radii)
rati_1000=fun_ratio(auto_1000,radii)

dist_array=np.array(range(200,1001,25))
rati_array=np.array([rati_0200,rati_0225,rati_0250,rati_0275,rati_0300,rati_0325,rati_0350,rati_0375,rati_0400,rati_0425,rati_0450,rati_0475,rati_0500,rati_0525,rati_0550,rati_0575,rati_0600,rati_0625,rati_0650,rati_0675,rati_0700,rati_0725,rati_0750,rati_0775,rati_0800,rati_0825,rati_0850,rati_0875,rati_0900,rati_0925,rati_0950,rati_0975,rati_1000])
plt.plot(dist_array,rati_array)
plt.title('Ratio [without subtracting mean]')
plt.title('Ratio [with: autocon=max(autocon,1)][with subtracting bckgrnd]')
# plt.title('Ratio [with: autocon=max(autocon,1)]')
plt.show()
```
