# Test 01: Measuring the relation between 'cm' and 'px'

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import math
from skimage import draw
from scipy import signal
from PIL import Image
```

## Step 1 - Analze one single LED

```python
#Import image
i1_00 = Image.open('from_camera/test01/t1d00.png')
realsize=i1_00.size
factor=4
newsize=tuple(int(ti/factor) for ti in realsize)
i1_00=i1_00.resize(newsize,Image.ANTIALIAS)
i1_00 = np.array(i1_00,dtype=float)
i1_00 = i1_00[:, :, 0] #For instance, at point (100,50) we had [0.608, 0.608, 0.608]. Now only have a single 0.608 value.

#Get rid of the offset
print('Mean:',i1_00.mean())
i1_00 = i1_00 - i1_00.mean()
print('Mean:',i1_00.mean())

#Print size
[size_y,size_x]=i1_00.shape
print(i1_00.shape) #[Y,X]
print('i1_00(t) imported and reduced correctly!')
print(size_x)
print(size_y)

#Print image
plt.imshow(i1_00,cmap='gray')
plt.show()
```

```python
#The filter: y0=h*x0=h
h=i1_00
```

```python
#Import image
i1_05 = Image.open('from_camera/test01/t1d05.png')
i1_05=i1_05.resize(newsize,Image.ANTIALIAS)
i1_05 = np.array(i1_05,dtype=float)
i1_05 = i1_05[:, :, 0]

#Print image
plt.imshow(i1_05,cmap='gray')
plt.show()


#Import image
i1_10 = Image.open('from_camera/test01/t1d10.png')
i1_10=i1_10.resize(newsize,Image.ANTIALIAS)
i1_10 = np.array(i1_10,dtype=float)
i1_10 = i1_10[:, :, 0]

#Print image
plt.imshow(i1_10,cmap='gray')
plt.show()


#Import image
i1_15 = Image.open('from_camera/test01/t1d15.png')
i1_15=i1_15.resize(newsize,Image.ANTIALIAS)
i1_15 = np.array(i1_15,dtype=float)
i1_15 = i1_15[:, :, 0]

#Print image
plt.imshow(i1_15,cmap='gray')
plt.show()

#Import image
i1_20 = Image.open('from_camera/test01/t1d20.png')
i1_20=i1_20.resize(newsize,Image.ANTIALIAS)
i1_20 = np.array(i1_20,dtype=float)
i1_20 = i1_20[:, :, 0]

#Print image
plt.imshow(i1_20,cmap='gray')
plt.show()
```

```python
hf=np.flip(h)
z1_00=signal.fftconvolve(i1_00,hf,mode='same')
z1_05=signal.fftconvolve(i1_05,hf,mode='same')
z1_10=signal.fftconvolve(i1_10,hf,mode='same')
z1_15=signal.fftconvolve(i1_15,hf,mode='same')
z1_20=signal.fftconvolve(i1_20,hf,mode='same')

#Print image
plt.imshow(z1_00,cmap='gray')
plt.show()
plt.imshow(z1_05,cmap='gray')
plt.show()
plt.imshow(z1_10,cmap='gray')
plt.show()
plt.imshow(z1_15,cmap='gray')
plt.show()
plt.imshow(z1_20,cmap='gray')
plt.show()
```
