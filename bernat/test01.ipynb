## Step 1 - Analyze one single LED

```python
#Import image
i1_00 = Image.open('from_camera/test01/t1d00.png')
realsize=i1_00.size
factor=4
newsize=tuple(int(ti/factor) for ti in realsize)
i1_00=i1_00.resize(newsize,Image.ANTIALIAS)
i1_00 = np.array(i1_00,dtype=float)
i1_00 = i1_00[:, :, 0] #For instance, at point (100,50) we had [0.608, 0.608, 0.608]. Now only have a single 0.608 value.

#Get rid of the offset
print('Mean:',i1_00.mean())
i1_00 = i1_00 - i1_00.mean()
print('Mean:',i1_00.mean())

#Print size
[size_y,size_x]=i1_00.shape
print(i1_00.shape) #[Y,X]
print('i1_00(t) imported and reduced correctly!')
print(size_x)
print(size_y)

#Print image
plt.imshow(i1_00,cmap='gray')
plt.show()
```

## Step 2 - Analyze two LEDs

```python
def sources_s(dist,sx,sy):
    output=np.zeros((sy,sx), dtype=float) #Generates a sy by sx null matrix
    output[math.floor(sy/2),math.ceil((sx-1)/2-dist)]=1  #Sets to 1 a position at distance -dist respect to the center
    output[math.floor(sy/2),math.floor((sx-1)/2+dist)]=1 #Sets to 1 a position at distance +dist respect to the center
    return output
```

```python
#The filter is the same
plt.imshow(h,cmap='gray')
plt.show()

#Import image

#05cm
i3_05 = Image.open('from_camera/test01/t3d05.png')
i3_05=i3_05.resize(newsize,Image.ANTIALIAS)
i3_05 = np.array(i3_05,dtype=float)
i3_05 = i3_05[:, :, 0]
plt.imshow(i3_05,cmap='gray')
plt.show()

#10cm
i3_10 = Image.open('from_camera/test01/t3d10.png')
i3_10=i3_10.resize(newsize,Image.ANTIALIAS)
i3_10 = np.array(i3_10,dtype=float)
i3_10 = i3_10[:, :, 0]
plt.imshow(i3_10,cmap='gray')
plt.show()

#15cm
i3_15 = Image.open('from_camera/test01/t3d15.png')
i3_15=i3_15.resize(newsize,Image.ANTIALIAS)
i3_15 = np.array(i3_15,dtype=float)
i3_15 = i3_15[:, :, 0]
plt.imshow(i3_15,cmap='gray')
plt.show()
```

```python
d_array=np.array([],dtype=float) #To make a plot, we initialize this np.array for the x-axis
norm_array=np.array([],dtype=float) #To make a plot, we initialize this np.array for the y-axis

range_d=math.floor(len(h[0])/2)
print(range_d) #We print the range to know, in advance, the number of iterations

for d in range(range_d): #To make things easier, range only form 5 to 16
    x_s=sources_s(d,size_x,size_y)  #We generate the possible input
    i_s=signal.fftconvolve(x_s, h,mode='same') #We filter it with h(t)
    norma = np.linalg.norm(i3_05-i_s) #We obtain the NORM
    d_array=np.insert(d_array,len(d_array),d) #X-array
    norm_array=np.insert(norm_array,len(norm_array),norma) #Y-array

#Optimization:
norm_minim=np.amin(norm_array)
min_pos,=np.where(norm_array==norm_minim)
d_minim=float(d_array[min_pos])

print('minimum norm:',norm_minim)
print('minimmum distance (px):',d_minim)
print('minimmum distance (cm):',d_minim/7.46)

#Plotting
plt.plot(d_array,norm_array)
plt.show()
```
